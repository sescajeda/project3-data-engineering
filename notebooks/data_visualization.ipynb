```python
# data_visualization.ipynb

# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned datasets (if you've saved them with encoding)
airbnb_data = pd.read_csv('../data/cleaned_airbnb_data.csv', encoding='ISO-8859-1')
hotel_data = pd.read_csv('../data/cleaned_hotel_data.csv', encoding='ISO-8859-1')

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Visualizing price distribution of Airbnb listings
plt.figure(figsize=(12, 6))
sns.histplot(airbnb_data['price'], bins=30, kde=True)
plt.title('Price Distribution of Airbnb Listings')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Visualizing star ratings for hotel data
plt.figure(figsize=(12, 6))
sns.countplot(x='star_rating', data=hotel_data, palette='viridis')
plt.title('Count of Hotels by Star Rating')
plt.xlabel('Star Rating')
plt.ylabel('Count of Hotels')
plt.show()
```
